```python
import pandas as pd
import numpy as np
```

```python
file = "C:\Learn\Python\Pandas Homework\School info.csv"
```

```python
file1 = "C:\Learn\Python\Pandas Homework\student name.csv"
```

```python
school_df = pd.read_csv(file)
school_df = school_df.set_index("school_name")

student_df = pd.read_csv(file1)
school_group_df = student_df.groupby("school_name")

student_df["reading_passed"] = student_df["reading_score"] >= 70
student_df["math_passed"] = student_df["math_score"] >= 70

school_df["budget_per_student"] = school_df["budget"] / school_df["size"]

school_group_df = student_df.groupby("school_name")

read_avg_df =  school_group_df["reading_score"].mean()
read_avg_df.columns = ["school_read_score_avg"]

math_avg_df =  school_group_df["math_score"].mean() 
math_avg_df.columns = ["school_math_score_avg"]

school_df["school_read_avg"] = school_group_df["reading_score"].mean()
school_df["school_math_avg"] = school_group_df["math_score"].mean()

school_df["school_math_pass_count"] = student_df.loc[student_df["math_score"] >= 70].groupby("school_name")["Student ID"].count()
school_df["school_read_pass_count"] = student_df.loc[student_df["reading_score"] >= 70].groupby("school_name")["Student ID"].count()

school_df["school_read_pass_pct"] = school_df["school_read_pass_count"] / school_df["size"] 
school_df["school_math_pass_pct"] = school_df["school_math_pass_count"] / school_df["size"]

school_df["school_overall_pass_pct"] = ( school_df["school_read_pass_pct"]  + school_df["school_math_pass_pct"]) / 2
top_school_df = school_df.sort_values (by="school_overall_pass_pct", ascending = False, inplace = False).head(5).copy()
bottom_school_df = school_df.sort_values (by="school_overall_pass_pct", ascending = True, inplace = False).head(5).copy()

```

```python
merged_df = pd.merge(student_df,school_df, on="school_name", how = "left")
group_by_school_type = merged_df.groupby("type")

score_by_school_type_df = group_by_school_type.mean()[["reading_score","math_score"]]

count_by_school_type = group_by_school_type.count()["Student ID"]

reading_pass_pct_by_school_type = merged_df[ merged_df["reading_passed"]== True ].groupby("type").count()["Student ID"] / count_by_school_type
math_pass_pct_by_school_type = merged_df[ merged_df["math_passed"]== True ].groupby("type").count()["Student ID"] / count_by_school_type

score_by_school_type_df["read_pass_pct"] = reading_pass_pct_by_school_type
score_by_school_type_df["math_pass_pct"] = math_pass_pct_by_school_type
score_by_school_type_df["overall_pass_pct"] = (score_by_school_type_df["read_pass_pct"] + score_by_school_type_df["math_pass_pct"] ) / 2

score_by_school_type_df

```
